# **Zadatak 2**

a)

```python
data = pd.read_csv('/content/drive/MyDrive/VI Content/izvjestaj.csv')
```

b)

```python
from sklearn.impute import SimpleImputer

si = SimpleImputer(strategy='median')

data.replace('/', np.nan, inplace=True)

print('Prije zamjene: ')
print(data.loc[:5, ['Ispit1','Ispit2']])

data[['Ispit1', 'Ispit2']] = si.fit_transform(data[['Ispit1', 'Ispit2']])

print('Nakon zamjene: ')
print(data.loc[:5, ['Ispit1','Ispit2']])
```

c)

```python
si = SimpleImputer(strategy='mean')

print('Prije zamjene: ')
print(data.loc[:5, ['Ispit1_popravni','Ispit2_popravni']])

data[['Ispit1_popravni', 'Ispit2_popravni']] = si.fit_transform(data[['Ispit1_popravni', 'Ispit2_popravni']])

print('Nakon zamjene: ')
print(data.loc[:5, ['Ispit1_popravni','Ispit2_popravni']])
```

d)

```python
from sklearn.preprocessing import scale

data[['Ispit1', 'Ispit2']] = scale(data[['Ispit1', 'Ispit2']])

data.loc[:5, ['Ispit1','Ispit2']]
```

e)

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()

data[['Ispit1_popravni', 'Ispit2_popravni']] = mm.fit_transform(data[['Ispit1_popravni', 'Ispit2_popravni']])

data.loc[:5, ['Ispit1_popravni','Ispit2_popravni']]
```

f)

```python
data.fillna(0, inplace=True)
data.head()
```

g)

```python
ocjene = data.pop('Ocjena')
ocjene.head()
```

```python
data.head()
```

h)

```python
data_np = data.values

ocjene_np = ocjene.values
ocjene_np = ocjene_np.astype(int)
print("NumPy niz originalnog skupa podataka:")
print(data_np[:5])

print("\nNumPy niz ocjena:")
print(ocjene_np[:5])
```

i)

```python
from sklearn.model_selection import train_test_split

X = data_np
Y = ocjene_np

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Y train", Y_train)
print("Y test", Y_test)
```
