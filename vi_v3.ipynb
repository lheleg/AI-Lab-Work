```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target

feature_names = iris.feature_names
target_names = iris.target_names

print("Nazivi znacajki:", feature_names)
print("Nazivi labela:", target_names)
print("\nPrvih 5 redova X:\n", X[:5])
```

b)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size = 0.3, random_state = 1
)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Analiza:

Koja funkcija je korištena za podjelu skupa podataka? train_test_split iz sklearn.model_selection

Koje parametre prima ova funkcija? ulazni podaci, izlazni podaci, velicina testnog skupa i seed za generator slucajnih brojeva

Koji je oblik novonastalih skupova?

*    X_train - matrica: 105 redova (instanci), 4 kolone (znacajke)
*    X_test - matrica: 45 redova (instanci), 4 kolone (znacajke)
*    y_train - niz ciljnih vrijednosti za trening od 105 elemenata
*    y_test - niz ciljnih vrijednosti za testiranje od 45 elemenata

Da li je ovo veliki skup podataka? U općem smislu, skup od 150 instanci nije ogroman, posebno za duboko učenje ili slične probleme koji zahtijevaju veliku količinu podataka.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

classifier_knn = KNeighborsClassifier(n_neighbors = 3)

classifier_knn.fit(X_train, y_train)
```

Na koji način se učitava k-NN? klasa KNeighborsClassifier iz sklearn.neighbors

Koja je vrijednost k? 3

d)

```python
y_pred = classifier_knn.predict(X_test)

print("Tacnost:", metrics.accuracy_score(y_test, y_pred))
```

# **Zadatak 4**

a)

```python
from keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print("Broj slika u trening skupu:", x_train.shape[0])
print("Broj slika u testnom skupu:", x_test.shape[0])
print("Shape varijable x_train:", x_train.shape)
print("Shape varijable x_test:", x_test.shape)
print("Shape varijable y_train:", y_train.shape)
print("Shape varijable y_test:", y_test.shape)
print("Veličina varijable x_train u bajtovima:", x_train.nbytes)
print("Veličina varijable x_test u bajtovima:", x_test.nbytes)
print("Veličina varijable y_train u bajtovima:", y_train.nbytes)
print("Veličina varijable y_test u bajtovima:", y_test.nbytes)
```

b)

```python
train_images = x_train.reshape((x_train.shape[0], 28 * 28))
train_images = train_images.astype('float32') / 255

test_images = x_test.reshape((x_test.shape[0], 28*28))
test_images = test_images.astype('float32') / 255
```

```python
print("Shape varijable x_train:", train_images.shape)
print("Shape varijable x_test:", test_images.shape)

print("Veličina varijable x_train u bajtovima:", train_images.nbytes)
print("Veličina varijable x_test u bajtovima:", test_images.nbytes)
```

1. Šta radi naredba reshape, a šta astype?

Reshape mijenja oblik, u datom primjeru mijenja trodimenzionalni oblik slike u dvodimenzionalni. Astype mijenja tip podataka, u datom primjeru broj piksela u vrijednosti raspona od 0 do 1.

2. Koliko bajta u memoriji sada zauzimaju slike za testiranje i trening?

Zauzimaju znantno vise.


c)

```python
from keras import models
from keras import layers

network = models.Sequential()
network.add(layers.Dense(512, activation='relu', input_shape=(28 * 28,)))
network.add(layers.Dense(10, activation='softmax'))
```

1. Pokušajte istražiti Keras dokumentaciju, te se upoznati sa dense slojem. Koje sve parametre prima ovaj sloj?

To su: units, activation, use_bias, kernel_initializer, bias_initializer, kernel_regularizer, bias_regularizer, activity_regularizer, kernel_constraint, bias_constraint, lora_rank

2. Na koji način je on povezan sa narednim slojem, te zašto za ove slojeve kažemo da su potpuno povezani?

Svaki neuron u dense sloju ima vezu sa svakim neuronom u sljedećem sloju. Zbog ovakvog nacina povezanosti, ovi slojevi se nazivaju potpuno povezanim slojevima.

d)

```python
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
```

e)

```python
from keras.utils import to_categorical

train_labels = to_categorical(y_train)
test_labels = to_categorical(y_test)
```

Funkcija to_categorical iz Kerasa koristi se za pretvaranje vektora klasa (npr. [0, 1, 2, 3]) u binarnu reprezentaciju tih klasa (npr. za klasu 2, [0, 0, 1, 0]). Njeni parametri su: y - ulazni vektor klasa koji se pretvara u binarnu reprezentaciju i num_classes - opcioni argument koji oznacava ukupan broj klasa. Povratna vrijednost funkcije je numpy array koji predstavlja binarnu reprezentaciju ulaznog vektora klasa.

f)

```python
network.fit(train_images, train_labels, epochs=5, batch_size=128)
```

epochs - oznacava broj prolazaka kroz cijeli trening skup podataka tokom treniranja neuronske mreže.

batch_size - predstavlja broj uzoraka koji se koristi za jedno azuriranje modela tokom treninga.

g)

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)

print('Tacnost za testni skup:', test_acc)
```

1. Kolika je tačnost nad testnim skupom podataka?

Cca 97.88%.

2. Da li se drastično razlikuje od tačnosti nad trening skupom?

Ne razlikuje se drasticno, sto je dobar pokazatelj.

3. Da li je ovo zadovoljavajući rezultat? Zašto jeste/nije?

Mozemo reci da je zadovoljavajuci rezultat, shodno vrijednosti i ponasanju tacnosti.
